# **Importing required libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

# **Load the Titanic dataset**

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

# **Display first few rows**

```python
print("ðŸ”¹ First 5 Rows:")
print(df.head())
```

```python
print("\nðŸ”¹ Dataset Info:")
print(df.info())
print("\nðŸ”¹ Summary Statistics:")
print(df.describe())
```

# **Checking for missing values**

```python
print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())
```

# **Visualizing missing data**

```python
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

# **Distribution of Numerical Features**

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols].hist(figsize=(10, 8), bins=20, color='skyblue')
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

# **Boxplots to detect outliers**

```python
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x=col)
    plt.title(f"Boxplot of {col}")
    plt.show()
```

# **Correlation matrix**

```python
# Select only numeric features for correlation analysis
numerical_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

# **Pairplot for selected features**

```python
selected_features = ['Survived', 'Pclass', 'Age', 'Fare']
sns.pairplot(df[selected_features], hue='Survived')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

# **Categorical Feature Analysis**

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, palette='pastel')
    plt.title(f"Countplot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

# Age distribution by **survival**

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x="Age", hue="Survived", fill=True)
plt.title("Age Distribution by Survival")
plt.show()
```

# Fare vs Survival - **interactiv**

```python
fig = px.box(df, x="Survived", y="Fare", color="Survived", points="all", title="Fare vs Survival")
fig.show()
```
